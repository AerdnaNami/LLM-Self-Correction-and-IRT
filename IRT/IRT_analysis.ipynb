```python
import pandas as pd 

file_path = 'total_naep_complete_2.csv' 
df = pd.read_csv(file_path)
df
```

```python
import numpy as np

gpt_df = pd.read_csv('responses_mcq/gpt_4o_naep_2.csv')
r1_df = pd.read_csv('responses_mcq/r1_naep_2.csv')
gemma3_df = pd.read_csv('responses_mcq/gemma3_naep_2.csv')
phi4_df = pd.read_csv('responses_mcq/phi4_naep_2.csv')
llama3_df = pd.read_csv('responses_mcq/llama3_naep_2.csv')

gpt_correctness = np.array(gpt_df['correctness'].to_list())
r1_correctness = np.array(r1_df['correctness'].to_list())
gemma3_correctness = np.array(gemma3_df['correctness'].to_list())
phi4_correctness = np.array(phi4_df['correctness'].to_list())
llama3_correctness = np.array(llama3_df['correctness'].to_list())
```

```python
gpt_correctness = np.transpose(gpt_correctness)
r1_correctness = np.transpose(r1_correctness)
phi4_correctness = np.transpose(phi4_correctness)
gemma3_correctness = np.transpose(gemma3_correctness)
llama3_correctness = np.transpose(llama3_correctness)
```

```python
try:
    from girth import ability_mle, ability_map, ability_eap
except ImportError:
    from girth.unidimensional.dichotomous import ability_mle, ability_map, ability_eap

llm_responses = np.c_[gpt_correctness, r1_correctness, phi4_correctness, gemma3_correctness, llama3_correctness]
print(llm_responses.shape)
```

```python
item_discrimination = np.array(df['a'].values)
item_difficulty = np.array(df['b'].values)
```

```python
# Verify shapes match
n_items = llm_responses.shape[0]
n_llms = llm_responses.shape[1]
assert item_discrimination.shape == (n_items,), "Discrimination array shape mismatch!"
assert item_difficulty.shape == (n_items,), "Difficulty array shape mismatch!"
```

```python
print(f"Estimating theta for {n_llms} LLMs based on {n_items} items.\n")

# --- Maximum A Posteriori (MAP) Estimation ---
# Similar to MLE, but incorporates a prior distribution for theta (typically N(0,1)
# by default in girth) to regularize estimates and prevent infinities.
# It finds the mode (peak) of the posterior distribution.
try:
    theta_map = ability_map(llm_responses, item_difficulty, item_discrimination)
    print("Theta Estimates (MAP with default N(0,1) prior):")
    print(theta_map)
    print("-" * 30)
except Exception as e:
    print(f"MAP estimation failed: {e}")
    print("-" * 30)
```
